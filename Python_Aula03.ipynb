## **1 - Comando While**
O Comando WHILE é utilizado para criarmos laços de repetição no código. Ou seja,

image.png

 **Como usar?**


---
Laço de repetição definido:

---

```
i = 6
while i < 6:
    print(i)
    i += 1
```


---
Laço de repetição indefinido:

---


```
i = 0
while True:
    print('Estou preso')
    i +=1
    if i == 6:
        break
```
---

Caso não usar o break, o programa entrará em um looping infinito, ocupando espaço na memória até o programa travar.


````python
i = 0


while i <= 6:
    print(i)
    i+= 1

print("Agora continuo meu código")
````

#  **2 - Comando FOR**

O comando FOR também é utilizado para fazer lógica de laços de repetição

image.png

 **Primeiro jeito de fazer FOR**

---

```
for contador in range(início, parada, incremento):
    print(contador)

```



---
**Segundo jeito de fazer FOR**

---


```
lista = [1, 2, 3, 4]
for contador in lista:
    print(contador)
```
---

Caso não usar o break, o programa entrará em um looping infinito, ocupando espaço na memória até o programa travar.


````python
print("Digite -1 a qualquer momento para sair")

while True:
    idade = int(input("Digite uma idade: "))
    if idade == -1:
        break
    else:
        print("Idade = {}".format(idade))
````

# **3 - Exercícios**

**1 -** Faça um programa que verifica se o usuário digitou um número inteiro.
Caso tenha digitado, mostrar o número na tela.
Caso não tenha digitado, pedir para o usuário digitar novamente.

````python
while True:
    numero = input("Digite um número inteiro: ")
    if numero.isnumeric():
        print("Você digitou {}.".format(numero))
        break
    else:
        print("Entrada Inválida")
````

**2 -** Faça um programa de verificação de login e senha. O usuário pode errar a senha até 3 vezes antes do programa terminar informando que o acesso foi bloqueado.

````python
login = "natan"
senha = '123'
contador = 0

while True:
    contador += 1
    username = input("Login: ").strip()
    password = input("senha: ").strip()
    if login == username and senha == password:
        print("Acesso liberado")
        break
    else: 
        if contador <= 2:
            print("Login e/ou senha inválidos")
        else:
            print("Acesso bloqueado")
            break
````

**3 -** Desenvolva um gerador de tabuada, capaz de gerar a tabuada de qualquer número inteiro entre 1 a 10. O usuário deve informar de qual numero ele deseja ver a tabuada. A saída deve ser conforme o exemplo abaixo:
Tabuada de 5:
5 X 1 = 5
5 X 2 = 10
...
5 X 10 = 50

````python
numero = int(input("Entre com um número: "))
contador = 1
print("Tabuada de {}".format(numero))
while contador <= 10:
    print("{} x {} = {}".format(numero,contador,numero * contador))
    contador += 1
````

**4 -** Desenvolva uma calculadora de fatorial em que o usuário digita o número e o programa retorna seu fatorial. Exemplo de saída esperada:




4! = 24

Faça uma verificação se o usuário realmente digitou alguma coisa.

````python
numero = input("Entre com um número: ")
fatorial = 1
if numero:
    num = int(numero)
    while num > 1:
        fatorial *= num
        num -= 1
    print("{}! = {}".format(numero,fatorial))
else:
    print("Não digitou nada")
````

**5 -** Considere a lista a seguir como a quantidade de itens vendidos por cada vendedor de uma loja.
vendas = [100, 150, 1500, 2000, 120]
Caso o vendedor tenha vendido mais de 1000 unidades, ele ganha um bônus de 15% sobre seu salário.
Fazer um programa que informa quem conseguiu o bônus.

````python
vendas = [100, 150, 1500, 2000, 120]
vendedores = ["João", "Fabiano", "Flávio", "Soraia", "Miguel"]

for venda in vendas:
    if venda < 1000:
        continue
    else:
        indice = vendas.index(venda)
        print(f'{vendedores[indice]} bateu a meta')
````
